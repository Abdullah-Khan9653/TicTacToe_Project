```python
# Simple Sentiment Analysis Tool (GUI)

import tkinter as tk
from tkinter import messagebox
from textblob import TextBlob

# Sentiment check function
def analyze_sentiment():
    user_text = text_input.get("1.0", tk.END).strip()
    if not user_text:
        messagebox.showwarning("Warning", "Please enter some text!")
        return
    
    polarity = TextBlob(user_text).sentiment.polarity
    
    if polarity > 0:
        sentiment, color = "Positive ðŸ˜Š", "green"
    elif polarity < 0:
        sentiment, color = "Negative ðŸ˜ž", "red"
    else:
        sentiment, color = "Neutral ðŸ˜", "blue"
    
    result_label.config(text=f"Sentiment: {sentiment}", fg=color)

# Reset function
def reset_text():
    text_input.delete("1.0", tk.END)
    result_label.config(text="Sentiment: None", fg="black")

# GUI setup
root = tk.Tk()
root.title("Simple Sentiment Analysis Tool")
root.geometry("500x400")
root.resizable(False, False)

# Title
tk.Label(root, text="Sentiment Analysis Tool", font=("Helvetica", 18, "bold")).pack(pady=10)

# Text input
text_input = tk.Text(root, height=8, width=50, font=("Helvetica", 12))
text_input.pack(pady=10)

# Buttons
button_frame = tk.Frame(root)
button_frame.pack()
tk.Button(button_frame, text="Analyze Sentiment", font=("Helvetica", 12), command=analyze_sentiment).grid(row=0, column=0, padx=10)
tk.Button(button_frame, text="Reset", font=("Helvetica", 12), command=reset_text).grid(row=0, column=1, padx=10)

# Result label
result_label = tk.Label(root, text="Sentiment: None", font=("Helvetica", 14))
result_label.pack(pady=20)

root.mainloop()
```
